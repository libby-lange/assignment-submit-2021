```python
import nltk
```

```python
import numpy as np
```

```python
from nltk import FreqDist
from nltk.tokenize import word_tokenize
from nltk.corpus import inaugural
```

```python
import matplotlib.pyplot as plt
```

```python
from nltk.corpus import stopwords
```

```python
trump_inaug = inaugural.words('2017-Trump.txt')
obama_inaug = inaugural.words('2009-Obama.txt')
```

```python
stopwords = nltk.corpus.stopwords.words('english')
```

```python
new_stopwords = [',', ';', '.', ':', '-', "'", 'The', 'For', 'And']
stopwords.extend(new_stopwords)
```

```python
trump_inaug_clean = []
obama_inaug_clean = []

for word in trump_inaug:
    if not word in stopwords:
        trump_inaug_clean.append(word)
    else:
        pass
    
for word in obama_inaug:
    if not word in stopwords:
        obama_inaug_clean.append(word)
    else:
        pass
```

```python
trump_inaug_clean_dist = FreqDist(trump_inaug_clean)

trump_inaug_clean_dist.plot(20)
```

```python
obama_inaug_clean_dist = FreqDist(obama_inaug_clean)

obama_inaug_clean_dist.plot(20)
```

```python
future_words = ['tomorrow', 'future', 'ahead', 'vision', 'bright', 'forward', 'progress', 'new', 'advance', 'change']
past_words = ['past', 'before', 'back', 'again', 'old', 'return', 'restore', 'reclaim', 'regain', 'stay', 'same']
my_fileids = ['2009-Obama.txt', '2013-Obama.txt', '2017-Trump.txt']
```

```python
plt.figure(figsize = (10,10))

cfd_future_obamatrump = nltk.ConditionalFreqDist((specific_word, fileid[:4]) #give args
    for fileid in my_fileids
    for each_word in inaugural.words(fileid)
    for specific_word in future_words
    
    if each_word.lower().startswith(specific_word))
        
cfd_future_obamatrump.plot()

plt.figure(figsize = (10,10))

cfd_past_obamatrump = nltk.ConditionalFreqDist((specific_word, fileid[:4]) #give args
    for fileid in my_fileids
    for each_word in inaugural.words(fileid)
    for specific_word in past_words
    
    if each_word.lower().startswith(specific_word))
        
cfd_past_obamatrump.plot()
```

```python
obama_future_2009 = 0
obama_future_2013 = 0
obama_past_2009 = 0
obama_past_2013 = 0

for word in inaugural.words('2009-Obama.txt'):
    if word in future_words:
        obama_future_2009 += 1
    if word in past_words:
        obama_past_2009 += 1
    else:
        pass
    
for word in inaugural.words('2013-Obama.txt'):
    if word in future_words:
        obama_future_2013 += 1
    if word in past_words:
        obama_past_2013 += 1
    else:
        pass
    
#collects all future and past obama words in separate values
```

```python
trump_future_2017 = 0
trump_past_2017 = 0
    
for word in inaugural.words('2017-Trump.txt'):
    if word in future_words:
        trump_future_2017 += 1
    if word in past_words:
        trump_past_2017 += 1
    else:
        pass

#collects all future and past trump words in separate values
```

```python
x1 = [2009, 2013, 2017]
y1 = [obama_future_2009, obama_future_2013, trump_future_2017]
x2 = [2009, 2013, 2017]
y2 = [obama_past_2009, obama_past_2013, trump_past_2017]


plt.xticks(np.arange(min(x1), max(x1)+1, 4)) #sets x axis interval to just inaug years
plt.plot(x1, y1, label = "Future word use")
plt.plot(x2, y2, label = "Past word use")
plt.xlabel('Number of words used')
plt.ylabel('Year')
plt.title('Comparing Obama and Trump\'s word use')
plt.legend()
plt.show()
```

# Your analysis here

Many Trump-era political analysts discussed how Trump's presidency played on a desire to return to some loosely defined "glorious" American past. To see if this hypothesis was reflected in his inaugural speech, I first plotted out his 20 most commonly used words. This pointed more towards a focus on "America" and "American" more than anything, so I decided to plot Obama's 20 most commonly used words to give myself a reference point. Again, few obvious conclusions jumped out from these simple plots.

Instead, I decided to create my own set of keyword lists, one containing 'future words' and one containing 'past words.' For example, I defined 'change', 'ahead', and 'tomorrow' as future words, and 'remain', 'again', and 'back' as past words. I then compared both of Obama's inaugural speeches and Trump's inaugural speech, plotting conditional frequency distributions for both future and past words. At first glance, it looked like Trump did use more past-facing words, but I wanted to be 100% sure, so I then plotted their total uses of past and future words. From this graph it was quite clear that past-facing words skyrocketed in Trump's inaugural speech, while future-facing words declined.

I then wanted to confirm whether the linguistic differences might be shared by all incumbent vs. new presidents, so I compared with another 2-term Democrat followed by a Republican: President Clinton and President George W. Bush. Here, it is clear that there is little difference in past-facing language between Clinton and Bush. In fact, Clinton actually used a number of specific 'past' terms more frequently than Bush. However, Bush also used far fewer future words than Clinton. Thus, it would appear Trump truly was a relatively past-facing President, but not necessarily and "excessively" future-adverse Republican President, although more comparisons with more samples might give a clearer picture on the second conclusion.

One other interesting finding I noted was that across both comparisons, Democrats' use of the word 'new' far outstripped Republican presidents. It makes sense that the party that champions 'progressive' values would push for more new projects, ideas, and policies, but it was interesting to see that reflected in their word choices.

Use the NYTimes script from last week to analyze the keyword distributions of the day

```python
import requests
from bs4 import BeautifulSoup
```

```python
title_list = []

nyt = 'https://www.nytimes.com/section/todayspaper' 
response = requests.get(nyt) 
soup2 = BeautifulSoup(response.text)
titles = soup2.find_all('h2')

for t in titles:
    title_list.append(t.get_text()) #add each headline to an empty list variable
```

```python
del title_list[0:3] #gets rid of useless link titles
```

```python
del title_list[-1]
```

```python
del title_list[-1]
```

```python
newer_stopwords = ['?', 'It', 'Be', '’', '‘']
stopwords.extend(newer_stopwords)
```

```python
title_string = ' '.join(title_list) #converts from list to string so you can tokenize
print(title_string)
```

```python
title_words = word_tokenize(title_string) #splits string into individual words
```

```python
title_words_clean = []

for word in title_words:
    if not word.lower() in stopwords:
        title_words_clean.append(word)
    else:
        pass
```

```python
title_freq = FreqDist(title_words_clean)
title_freq.plot(20)
```

```python
bi_title = list(nltk.bigrams(title_words_clean))
bi_fdist_title = FreqDist(bi_title)
print(bi_fdist_title.plot(10))
```

```python
subtitle_list = []

base_url = 'http://www.nytimes.com'
r = requests.get(base_url)
soup = BeautifulSoup(r.text)
paragraphs = soup.find_all('p')

for p in paragraphs:
    subtitle_list.append(p.get_text())
```

```python
subtitle_string = ' '.join(subtitle_list)
subtitle_words = word_tokenize(subtitle_string)
```

```python
subtitle_words_clean = []

for word in subtitle_words:
    if not word.lower() in stopwords:
        subtitle_words_clean.append(word)
    else:
        pass
```

```python
subtitle_freq = FreqDist(subtitle_words_clean)
subtitle_freq.plot(20)
```
